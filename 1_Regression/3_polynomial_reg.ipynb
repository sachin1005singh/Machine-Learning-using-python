### Polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

```python
np.random.seed(42)
#This method is called when `RandomState` is initialized.
n_samples = 100

X = np.linspace(0,10,n_samples)
c = np.random.randn(100)

y = X**3+c*100+100


plt.scatter(X,y)
```

```python

print("x shape :",X.shape)
print("y : ",y.shape)
print("x dim = ",X.ndim,"y dim = ",y.ndim)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X.reshape(-1,1),y)

model_pred = lr.predict(X.reshape(-1,1))

plt.scatter(X,y)
plt.plot(X,model_pred)
print (r2_score(y,model_pred))
```

#### here linear regression is not able to well fitted all data points

### Polynomial Regression

```python
#Polynominal Regression:Transformer (not a model to make prediction)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures()
X=X.reshape(-1,1)
x_poly = poly_reg.fit_transform(X)
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
y_pred = lin_reg2.predict(x_poly)
```

```python
#plt.figure(figsize=(6,4))
plt.scatter(X,y)
plt.plot(X,y_pred)
print (r2_score(y,y_pred))
```
