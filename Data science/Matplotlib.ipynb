# Basic plotting: introduction to matplotlib

show() etc. pyplot is one of the most commonly used module for creating a variety of plots such as the line plots. bar plots, 
histogram etc.

# Figures and Subplots

using figues and subplot you can create multiple subplot inside the same figure.
To create multiple plot in the same figure, you can use ,method plt.subplot(nrows,ncols,nsubplot).

```python
x=np.linspace(1,10,100)
y=np.log(x)

plt.subplot(121)
#1 for row
#2 for column
#1 place of the figure

plt.title('y=log(x)')
plt.plot(x,y)

plt.subplot(122)
plt.title('y=log(x)**2')
plt.plot(x,y**2)
```

```python
x=np.linspace(1,10,100)
y=np.linspace(2,10,100)
d=np.log(x)
c=np.exp(x)

plt.figure(figsize=(8,8))
plt.subplot(221)
#1 for row
#2 for column
#1 place of the figure

plt.title('y=log(x)')
plt.plot(x,d)

plt.subplot(222)
plt.title('y=log(x)**2')
plt.plot(x,y**2)

plt.subplot(223)
plt.title('cube')
plt.plot(x,y**3)

plt.subplot(224)
plt.title('exponential')
plt.plot(x,c)

```

```python
x=np.linspace(1,10,100)
plt.figure(figure size(8,6))
plt.figure(1)

plt.subplot(221)
plt.title('Linear')
plt.plot(x,x)

plt.subplot(222)
plt.title('cubic')
plt.plot(x,x**3)
```
